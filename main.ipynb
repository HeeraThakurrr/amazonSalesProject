## Data Cleaning

```python
df.isnull().sum().sort_values(ascending=False)
```

### Drop useless columns and Fill null values

```python
df.drop(columns=['Unnamed: 22'], inplace=True)
```

```python
df['Amount'] = df['Amount'].fillna(0)
```

```python
df.dropna(subset=['ship-city', 'ship-state'], inplace=True)
```

```python
df['Courier Status'] = df['Courier Status'].fillna('Unshipped')
```

```python
df.isnull().sum()
```

# Overall Stats

```python
total_revenue = df['Amount'].sum()
total_orders = df['Order ID'].nunique()
avg_order_val = total_revenue/total_orders

print(f"Total Revenue: {total_revenue}")
print(f"No. of Orders: {total_orders}")
print(f"Average order value: {avg_order_val}")
```
